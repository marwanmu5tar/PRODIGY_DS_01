# Population Analysis and Visualization

## Introduction

This code aims to analyze and visualize population data for the top 20 countries. The analysis includes a bar plot displaying the population of each country and a violin plot depicting the distribution of ages by gender.

## Data Preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create some example data
data = {
    "Country Name": ["China", "India", "United States", "Indonesia", "Pakistan", "Brazil", "Nigeria", "Bangladesh", "Russia", "Mexico", "Japan", "Ethiopia", "Egypt", "Democratic Republic of the Congo", "Philippines", "Vietnam", "Germany", "Turkey", "Iran"],
    "Population": [1444, 1380, 330, 273, 234, 212, 206, 164, 146, 130, 126, 115, 111, 103, 99, 83, 85, 84, 90]
}

# Add random age and gender data
np.random.seed(42)
ages = np.random.randint(20, 60, size=len(data["Country Name"]))
genders = np.random.choice(["Male", "Female"], size=len(data["Country Name"]))

data["Age"] = ages
data["Gender"] = genders

# Create a pandas DataFrame from the data
df = pd.DataFrame(data)
```

```python
# Sort the data by population
df = df.sort_values(by='Population', ascending=False)

# Top 20 countries by population
top_countries = df.head(20)
```

# Bar Plot for Population


```python
# Set the seaborn style
sns.set(style="whitegrid")

# Create a bar plot using Seaborn for population
plt.figure(figsize=(14, 8), facecolor='#ecf0f1')
bars = sns.barplot(x='Country Name', y='Population', data=top_countries, palette='viridis', edgecolor='#2c3e50')

# Add labels on top of each bar
for bar, population in zip(bars.patches, top_countries['Population']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 5, f'{population:,}M', ha='center', va='bottom', fontsize=10, color='#2c3e50')

# Create a violin plot using Seaborn for age and gender distribution
plt.title('Top 20 Countries by Population', fontsize=20, fontweight='bold', color='#2c3e50')
plt.xlabel('Country', fontsize=16, color='#2c3e50')
plt.ylabel('Population (Millions)', fontsize=16, color='#2c3e50')
plt.xticks(rotation=45, ha='right', fontsize=12, color='#2c3e50')
plt.yticks(fontsize=12, color='#2c3e50')
plt.ylim(0, max(top_countries['Population']) + 50)  # Adjust y-axis limits for better visibility
plt.grid(axis='y', linestyle='--', alpha=0.5)


# Remove the top and right spines for a cleaner appearance
sns.despine(top=True, right=True)
```

# Bar Plot for Gender Distribution

```python
plt.figure(figsize=(12, 8))
sns.violinplot(x='Gender', y='Age', data=df, palette={"Male": "#3498db", "Female": "#e74c3c"}, inner='quartile')
plt.title('Distribution of Age by Gender', fontsize=20, fontweight='bold', color='#2c3e50')
plt.xlabel('Gender', fontsize=16, color='#2c3e50')
plt.ylabel('Age', fontsize=16, color='#2c3e50')
plt.xticks(fontsize=12, color='#2c3e50')
plt.yticks(fontsize=12, color='#2c3e50')
plt.grid(axis='y', linestyle='--', alpha=0.5)
# Remove the top and right spines for a cleaner appearance
sns.despine(top=True, right=True)
```
